```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
#Loading the data
data = pd.read_csv('amazon.csv')
```

```python
data
```

# Data Cleaning

```python
data.isnull().sum()
```

```python
data.dropna()
```

```python
#checking null values
data.isnull().sum()
```

```python
#Remove rows
data.dropna(subset=['rating_count'], inplace=True)
```

```python
data.info()
```

```python
#Remove column
data.drop('product_link',axis=1,inplace=True)
```

```python
#Duplicate check
data.duplicated().sum()
```

```python
print("Dataset remains as ", data.shape[0], " rows and ", data.shape[1], " columns")
```

```python
#Clean data header name
namechange={"product_id":"Product ID", "product_name":"Product Name", "category":"Category", "discounted_price":"Discounted Price", "actual_price":"Actual Price", "discount_percentage":"Discount Percentage", "rating":"Rating", "rating_count":"Rating Count", "about_product":"About Product", "user_id":"User ID", "user_name":"User Name", "review_id":"Review ID", "review_title":"Review Title", "review_content":"Review Content", "img_link":"IMG Link", "product_link":"Product Link"}
data=data.rename(columns=namechange)
```

```python
data
```

```python
#Replace symbols for relevant columns
data['Discounted Price']=data['Discounted Price'].str.replace("₹",'').str.replace(",",'')
data['Actual Price']=data['Actual Price'].str.replace("₹",'').str.replace(",",'')
data['Discount Percentage']=data['Discount Percentage'].str.replace("%",'')
data['Rating Count']=data['Rating Count'].str.replace(",",'')
```

```python
data
```

```python
data['Product ID'].value_counts()
```

```python
data['Product Name'].unique()
```

```python
#Resolve error for the conversion of Rating
unique_ratings=data['Rating'].unique()
print(unique_ratings)
data=data.loc[data['Rating'] != '|'] # remove column from dataset
```

```python
data
```

```python
#Convert data type to float
data=data.astype({'Discounted Price': 'float64', 'Actual Price': 'float64', 'Discount Percentage': 'float64', 'Rating': 'float64', 'Rating Count': 'float64'})
```

```python
data.info()
```

```python
grouped=data.groupby('Category')['Product ID'].apply(list).reset_index()
print(grouped)
```

```python
grouped=data.groupby('Category')['Product ID'].apply(lambda x: ','.join(x)).reset_index()
grouped
```

```python
#Count the number of products in each category
category_counts=data['Product ID'].value_counts().sort_index()
category_counts
```

# Data Visualization

```python
plt.figure(figsize=(10, 6))
category_counts.plot(kind='line', marker='o', color='blue')
plt.title('Number of Products in Each ID')
plt.xlabel('Product ID')
plt.ylabel('Number of Products')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Product ID'], data['Actual Price'], marker='o', color='blue', label='Actual Price')
plt.plot(data['Product ID'], data['Discounted Price'], marker='o', color='red', label='Discounted Price')
plt.title('Actual Price vs Discounted Price')
plt.xlabel('Product ID')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
data.columns
```

```python
#Sort the DataFrame by 'Rating' in descending order
df_sorted=data.sort_values(by='Rating', ascending=False)
```

```python
# elect the top N products with the highest ratings
top_n=5
top_rated_products=df_sorted.head(top_n)
```

```python
plt.figure(figsize=(13, 8))
plt.bar(top_rated_products['Product ID'], top_rated_products['Rating'], color='skyblue')
plt.title('Top Rated Products')
plt.xlabel('Product ID')
plt.ylabel('Rating')
plt.ylim(0, 5)  # Set the y-axis limit to the range of ratings (0 to 5)
plt.xticks(rotation=45)
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['Rating'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#Find the highest rating against each product
highest_ratings=data.groupby('Product ID')['Rating'].max().reset_index()
print(highest_ratings)
```
